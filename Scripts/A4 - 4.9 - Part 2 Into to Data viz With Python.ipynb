### 7 - Q5 - Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. Use sample data in line chart.

#### >> Creating a list holding True/False values to the test np.random.rant() <= 0.7

```python
# Creating a list holding True/False values to the test np.random.rant() <= 0.7

np.random.seed(4)

dev = np.random.rand(len(df_ord_prod_cust)) <= 0.7
```

#### >> Splitting the df in two parts - Big (70%) and Small (30%)

```python
# Storing 70% of the data into a BIG dataframe

big = df_ord_prod_cust[dev]
```

```python
# Storing 30% of the data into a SMALL dataframe

small = df_ord_prod_cust[~dev]
```

#### >> Checking if the total matches for two parts

```python
len(df_ord_prod_cust)
```

```python
len(small)+len(big)
```

#### >> Reducing the dataframe to have only specific columns needed for the analysis or a line chart

```python
# Creating containedsample (a subset from the 'small' sample group of 30% data) wiht only two columns

containedsample = small[['order_hour_of_day','prices']]
```

#### >> Creating a line chart using the small dataset with only related columns

```python
# Create a line chart

line_hour_price = sns.lineplot(data = containedsample, x = 'order_hour_of_day', y = 'prices')
```

#### >> Saving the line chart

```python
# saving the line chart

line_hour_price.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_hour_price.png'))
```

```python
containedsample['prices'].max()
```

```python
containedsample['prices'].min()
```

```python
containedsample['prices'].mean()
```

#### >> Observation

1. This vizualisation is based on the sample data set.
2. The sample is truly reflective of the larger data set. The maximum, minimum and mean values of the 'containedsample' which are inline with the larger dataset.
3. The most busy hours of the day around 10am to 3pm is the period where customers are buying most averagely priced items and even slightly lower than the average.
4. We can interpret that the large order volumn makes this the busiest time of the day but prices are average or below average.
