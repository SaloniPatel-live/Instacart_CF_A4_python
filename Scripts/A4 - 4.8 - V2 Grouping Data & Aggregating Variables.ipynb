### 8 - Aggregating Data with transform()

### To create your flag, you’ll need some criteria. You can use the following:

If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

let’s map this task onto the three-step process introduced earlier:

1. Split the data into groups based on the “user_id” column.
2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.

```python
# ords_prods_merge['max_order'] = >>>> creating a new column 

# ords_prods_merge.groupby(['user_id'])['order_number'] >>>> group by user_id and order_number. so it will show which user_id has how many orders

# .transform(np.max) >>> transform the data 

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check the newly added column using head

ords_prods_merge.head(10)
```

### creating flag in a column using loc function

```python
# creating the loyalty flag using the max_order value created in the earlier command

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checking the frequency 

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# checking the specific columns with head(). This allows for a retrival of slightly larger and specific set of data for a quick check.

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(60)
```

### 6 - Creating a spending flag

#### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

- If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
- If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

First we will have to created a new column of average price (spent) for each user_id and then we will use the above given criterias to create spending flags.

#### Creating new column for avg_prod_price and aggregating values by user_id

```python
# Aggregating Data with transform() Function to create a new column

ords_prods_merge['avg_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Lets check out the newly created column along with user_id. I am using drop_duplicate function to only show unique values as without it I had many duplicates in my initial run

ords_prods_merge[['user_id', 'avg_prod_price', 'max_order']].drop_duplicates().head(15)
```

#### Creating 'spending flags'

```python
# Creating a 'Spending' Label Flag using loc 
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Value Count on newly created 'spending_flag' Column

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 7 - Creating Order Frequency Flags

#### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating a new volumn median_days_betn_orders

ords_prods_merge['median_days_betn_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Lets check the new column

ords_prods_merge[['user_id', 'days_since_last_order', 'median_days_betn_orders']].head(15)
```

```python
# Creating a 'Order Frequency' Label Flag using loc
ords_prods_merge.loc[ords_prods_merge['median_days_betn_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_betn_orders'] <= 20) & (ords_prods_merge['median_days_betn_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_betn_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Value Counts on 'order_frequency_flag'

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking newly created columns

ords_prods_merge[['user_id', 'days_since_last_order', 'median_days_betn_orders', 'order_frequency_flag']].head(15)
```
