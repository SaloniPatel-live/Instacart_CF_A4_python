### 6 - Creating a new user defined variable 'price_label' using if-elif 

```python
# Creating a new user defined variable 'price_label' 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

### 7 - Creating a new column called price_range. Values will be filled in using the variable 'price_label'

```python
# Creating a new column called price_range. Values will be filled in using the variable 'price_label' - in df

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# looking for the most expensive price - in df

df['prices'].max()
```

### 8 - loc function instead of it-elif - on df - another way to avoid doing the user defined function

```python
# testing loc function instad of if-elif statement - on df

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

#### NOTE >> output from both user defined function of 'price_range' and loc fuction of 'price_range_loc' are the same

### 4 - When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#Checking the frequecny of the values in variable order_day_of_week

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

#### Note >> 

1. 9am to 4:59pm = most orders
2. 7am to 8:59am ; 5pm to 10:59pm = Average orders
3. 11pm to 6:59am = Fewest orders

   This information is useful to create the next statement and assign them into a particular group.

```python
# Creating a function to assign 'Most orders', 'Average orders' and 'Fewest orders' labels for each order following the corresponding value of 'orders_hour_of_day'

result2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in (9, 10, 11, 12, 13, 14, 15, 16):
        result2.append("Most orders")
    elif value in (7, 8, 17, 18, 19, 20, 21, 22):
        result2.append("Average orders")
    else:
        result2.append("Fewest orders")
```

```python
# Creating the column 'busiest_period_of_day' and values stored in 'result2' will be assigned to this.

df_ords_prods_merge['busiest_period_of_day'] = result2
```

### 5 - Print the frequency for this new column.

```python
# Checking the value frequency for the newly created column

df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
