#### >> There are various ways to tackle this question. There are multiple answers.

1. Creating a user profile based on age, income, items in orders, and number of dependents will require certain amount of assumptions and educated guess.
2. Profiles could be as follow:
- SA - Single adult -- age > 18, dependents = 0, income > 28,000
- LISA - Low Income Single adult -- age > 18, dependents = 0, income < 28,000
- YP - Young parent -- age between 18 to 25, dependent 1 or more, income > 28,000
- HISA - High income single adult -- age > 18, dependents = 0 , income > 200,000
- HIFK - High income Family with kids -- age >  age > 18, dependents = 1 or more , income < 200,000
- LIFK - Low income Family with kids -- age >  age > 18, dependents = 1 or more , income < 100,000
- MIFK - Mid income Family with kids -- age >  age > 18, dependents = 1 or more , income between 100,000 and 200,000
- RET - Retirees -- age > 58, dependents = 0, income > 70,000
- elif age >= 18 and dependents == 0 and income < 28000:
        return 'LISA'
3. I will be creating a coulmn in main df_ord_prod_cust to indicate these profile ids

Assumption - 
1. National minimum rate of $14 per hour, an adult working for 8 hours a day, 5 days a week, 50 weeks a year pre tax
2. YP, HIFK, LIFK and MIFK may be shopping from 'babies' category 
3. HISF, HISM - spending on personal care, international food and alchohol
4. HISA - spending on personal care, international food and alchohol


#### >> Creating profiles

```python
# Creating profiles. This profile classification includes every single customer in the df

def assign_user_profile(row):
    age = row['age']
    income = row['income']
    dependents = row['dependents']

    if age >= 58 and dependents == 0 and income > 70000:
        return 'RET' #retired
    elif age >= 18 and dependents == 0 and income > 200000:
        return 'HISA'
    elif age >= 18 and dependents == 0 and income > 28000:
        return 'SA'
    elif age >= 18 and dependents == 0 and income < 28000:
        return 'LISA'
    elif 18 <= age <= 25 and dependents >= 1 and income > 28000:
        return 'YP'
    elif age >= 18 and dependents >= 1:
        if income < 100000:
            return 'LIFK'
        elif 100000 <= income < 200000:
            return 'MIFK'
        else:
            return 'HIFK'
    else:
        return 'Unclassified'
```

```python
# Applying the profiles under column 'user_profile' in the df_ord_prod_cust

df_ord_prod_cust['user_profile'] = df_ord_prod_cust.apply(assign_user_profile, axis=1)
```

```python
# Checking the value count of the user_profile. Also adding the counts into a variable called profile_count and printing it. There are no unclassified user profiles. 

profile_counts = df_ord_prod_cust['user_profile'].value_counts()

print(profile_counts)
```

#### >> Aggregate the max, mean, and min variables on a customer-profile level for usage frequency. 

We are unable to use the order_frequency_flag for this purpose. We need to use the actual order_id and count function.

```python
# Step 1 - Count number of orders per user_id, and also show the related user_profile

order_counts = df_ord_prod_cust.groupby(['user_id', 'user_profile'])['order_id'].count().reset_index(name='order_count')
```

```python
# Step 2 - lets look at the order_counts that we just created

order_counts
```

```python
# Step 3: Calculate min, mean, and max of these order_counts

order_summary = order_counts.groupby('user_profile')['order_count'].agg(['min', 'mean', 'max'])
```

```python
order_summary
```

```python
# Above table is perfect representation of the information but Lets Create a visualisation of stacked bar chart for this

# Prep data for stacking
order_summary = order_summary.reset_index()
profiles = order_summary['user_profile']
min_vals = order_summary['min']
mean_vals = order_summary['mean']
max_vals = order_summary['max']

# Plot stacked bars
plt.figure(figsize=(10,6))
plt.bar(profiles, min_vals, label='Min')
plt.bar(profiles, mean_vals, bottom=min_vals, label='Mean')
plt.bar(profiles, max_vals, bottom=min_vals + mean_vals, label='Max')

plt.title('Stacked Order Count Stats per User Profile')
plt.xlabel('User Profile')
plt.ylabel('Order Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### >> Comparing customer profiles with regions and departments - Average spend, Total Spend and Order Count
